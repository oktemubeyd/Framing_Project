```python
import pandas as pd
import numpy as np
from sklearn.metrics import cohen_kappa_score
from statsmodels.stats.inter_rater import fleiss_kappa
import krippendorff
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_annotations(file_path):
    """Load annotation data from an Excel file."""
    return pd.read_excel(file_path)
```

```python
def calculate_agreement_metrics(df, annotator1, annotator2):
    """Calculate agreement metrics between two annotators."""

    # Define invalid/missing placeholder values
    invalid_values = ["?", "?1", "1?", "", None]

    # Filter out rows where either annotator has an invalid value
    df = df[[annotator1, annotator2]].copy()
    df = df[~df[annotator1].isin(invalid_values)]
    df = df[~df[annotator2].isin(invalid_values)]
    
    # Percentage Agreement
    agreement = (df[annotator1] == df[annotator2]).mean()
    
    # Cohen's Kappa
    kappa = cohen_kappa_score(df[annotator1], df[annotator2])
    
    return {
        "Percentage Agreement": agreement,
        "Cohen's Kappa": kappa
    }
```

```python
def visualize_disagreement(df, annotator1, annotator2):
    """Plot a confusion matrix to visualize disagreement patterns."""
    confusion_matrix = pd.crosstab(df[annotator1], df[annotator2], rownames=[annotator1], colnames=[annotator2])
    plt.figure(figsize=(6,5))
    sns.heatmap(confusion_matrix, annot=True, cmap='Blues', fmt='d')
    plt.title('Disagreement Heatmap')
    plt.show()
```
